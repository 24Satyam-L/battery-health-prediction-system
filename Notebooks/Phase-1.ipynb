```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Reading CSV file into DataFrame
Dis_df = pd.read_csv(rf"C:\Projects\Battery Engineering\data\00001.csv")
Chg_df = pd.read_csv(rf"C:\Projects\Battery Engineering\data\00003.csv")

```

```python
# Displaying Information Related to Discharging Cycle
print("Discharging Cycle Data:")
print(Dis_df.info())
print(Dis_df.head())
Dis_df.describe()
```

```python
# Displaying Information Related to Charging Cycle
print("\nCharging Cycle Data:")
print(Chg_df.info())
print(Chg_df.head())
Chg_df.describe()
```

Viewing Plots Of Voltage vs Time for Charging and Discharging Cycle

```python
Dis_df.plot(x="Time",y="Voltage_measured")
Chg_df.plot(x="Time",y="Voltage_measured")
```

```python
Dis_df.plot(x='Time', y='Current_measured')
Chg_df.plot(x='Time', y='Current_measured')

```

Finding Sampling Frequency

```python
Dis_df['Delta_Time'] = Dis_df['Time'].diff()
Chg_df['Delta_Time'] = Chg_df['Time'].diff()

print(Dis_df['Delta_Time'].describe())
print(Chg_df['Delta_Time'].describe())
```
